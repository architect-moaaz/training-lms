# Python Data Structures

Welcome to Day 2! Today we'll explore Python's built-in data structures.

## Dictionaries

Dictionaries store key-value pairs.

```python
# Create a dictionary
student = {
    "name": "John",
    "age": 20,
    "major": "Computer Science"
}

print("Student:", student)
print("Name:", student["name"])
print("Age:", student["age"])
```

## Sets

Sets are unordered collections of unique items.

```python
# Create a set
colors = {"red", "green", "blue", "red"}
print("Colors:", colors)  # Note: duplicate 'red' is removed

# Add to set
colors.add("yellow")
print("Updated colors:", colors)
```

## Tuples

Tuples are immutable sequences.

```python
# Create a tuple
coordinates = (10, 20)
print("Coordinates:", coordinates)
print("X:", coordinates[0])
print("Y:", coordinates[1])
```

## Working with Lists

Let's explore more list operations.

```python
numbers = [1, 2, 3, 4, 5]

# Append
numbers.append(6)
print("After append:", numbers)

# Sum
total = sum(numbers)
print("Sum:", total)

# Max and Min
print("Max:", max(numbers))
print("Min:", min(numbers))
```
